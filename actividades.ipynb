## Actividad 8 — Bucle `for`: recorrer un rango

El bucle **`for`** recorre una secuencia de valores. Con `range(inicio, fin, paso)` generamos rangos de números.

**Consigna:** Utiliza un bucle `for` con `range()` para mostrar:
1. Los números del 0 al 9
2. Los números pares entre 0 y 20
3. Una cuenta regresiva del 10 al 1

**Ejemplo de salida esperada (punto 3):**
```
10
9
8
...
1
```

````python
# Escribí tu código acá
print("====================================")
print("Alumno: Zigaran Lucas Natanael")
print("====================================")
print("1. Los numeros del 0 al 9")
for i in range(10):
    print (i)

print("2. Los numeros pares entre 0 y 20")
for i in range(0,21,2):
    print(i)

print("3.Una cuenta regresiva del 10 al 1")
for i in range(10,0,-1):
    print(i)
````

## Actividad 10 — Desafío integrador

Actividad final. Esta combina todo lo que aprendiste.

**Consigna:** Crea un programa que:

1. Le pida al usuario cuántos números quiere ingresar.
2. Use un bucle para que el usuario ingrese esa cantidad de números.
3. Al finalizar, muestre:
   - El **mayor** número ingresado
   - El **menor** número ingresado
   - La **suma** de todos
   - El **promedio**
   - Cuántos números **pares** e **impares** se ingresaron

**Ejemplo de salida esperada:**
```
Mayor: 98
Menor: 3
Suma: 256
Promedio: 32.0
Pares: 4
Impares: 4
```


````python
print("====================================")
print("Alumno: Zigaran Lucas Natanael")
print("====================================")
cantidad = int(input("¿cuantos números desea ingresar?"))

mayor = float('-inf')    
menor = float('inf')       
suma = 0                   
pares = 0                  
impares = 0                

for i in range(cantidad):
    numero = int(input(f"Ingrese el número {i+1}: "))
    
    if numero > mayor:
        mayor = numero
    
    if numero < menor:
        menor = numero
    
    suma += numero
    
    if numero % 2 == 0:
        pares += 1
        print(f"{numero} es PAR")
    else:
        impares += 1
        print(f"{numero} es IMPAR")

promedio = suma / cantidad

print("RESULTADOS")
print(f"Números ingresados: {cantidad}")
print(f"Número mayor: {mayor}")
print(f"Número menor: {menor}")
print(f"Suma total: {suma}")
print(f"Promedio: {promedio:.2f}")  
print(f"Números pares: {pares}")
print(f" Números impares: {impares}")
````
